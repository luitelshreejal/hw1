```python

def mean(values):
    return sum(values) / len(values)

def variance(values, mu):
    return sum((x - mu) ** 2 for x in values) / len(values)

def covariance(x, y, mu_x, mu_y):
    return sum((xi - mu_x) * (yi - mu_y) for xi, yi in zip(x, y)) / len(x)

def correlation(x, y, mu_x, mu_y, var_x, var_y):
    return covariance(x, y, mu_x, mu_y) / math.sqrt(var_x * var_y)

M = 10000
U1 = [random.uniform(0, 1) for _ in range(M)]
U2 = [random.uniform(0, 1) for _ in range(M)]

Z1 = [math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2) for u1, u2 in zip(U1, U2)]
Z2 = [math.sqrt(-2 * math.log(u1)) * math.sin(2 * math.pi * u2) for u1, u2 in zip(U1, U2)]

mean_Z1, mean_Z2 = mean(Z1), mean(Z2)
var_Z1, var_Z2 = variance(Z1, mean_Z1), variance(Z2, mean_Z2)
corr_Z1_Z2 = correlation(Z1, Z2, mean_Z1, mean_Z2, var_Z1, var_Z2)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(Z1, bins=50, density=True, alpha=0.7, color='blue', label="Z1")
plt.title("Histogram of Z1 (Standard Normal)")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(Z2, bins=50, density=True, alpha=0.7, color='green', label="Z2")
plt.title("Histogram of Z2 (Standard Normal)")
plt.legend()

plt.tight_layout()
plt.show()

theoretical_mean_Z1_Z2 = 0  # For standard normal distribution, the theoretical mean is always 0
theoretical_variance_Z1_Z2 = 1  # For standard normal distribution, the theoretical variance is always 1

results_theoretical_empirical = {
    "Statistic": ["Mean", "Variance", "Correlation"],
    "Theoretical Z1, Z2": [theoretical_mean_Z1_Z2, theoretical_variance_Z1_Z2, 0],
    "Empirical Z1": [mean_Z1, var_Z1, None],
    "Empirical Z2": [mean_Z2, var_Z2, None],
    "Empirical Correlation Z1, Z2": [None, None, corr_Z1_Z2]
}

pd.DataFrame(results_theoretical_empirical)
```

```python
import matplotlib.pyplot as plt

# Function for Cholesky decomposition
def cholesky_decomposition(matrix):
    n = len(matrix)
    L = [[0.0] * n for _ in range(n)]

    for i in range(n):
        for j in range(i + 1):
            sum_val = sum(L[i][k] * L[j][k] for k in range(j))

            if i == j:
                L[i][j] = (matrix[i][i] - sum_val) ** 0.5
            else:
                L[i][j] = (matrix[i][j] - sum_val) / L[j][j]

    return L

rho = 0.5
mu_X, mu_Y = 1, -1
var_X, var_Y = 1, 4
sigma_X, sigma_Y = var_X ** 0.5, var_Y ** 0.5
M = 10000


cov_XY = rho * sigma_X * sigma_Y
cov_matrix = [[var_X, cov_XY], [cov_XY, var_Y]]

L = cholesky_decomposition(cov_matrix)

Z = [[random.gauss(0, 1) for _ in range(2)] for _ in range(M)]

XY = [[sum(L[i][k] * Z[m][k] for k in range(2)) for i in range(2)] for m in range(M)]


for i in range(M):
    XY[i][0] += mu_X
    XY[i][1] += mu_Y


X = [row[0] for row in XY]
Y = [row[1] for row in XY]


empirical_mean_X = sum(X) / M
empirical_mean_Y = sum(Y) / M
empirical_var_X = sum((x - empirical_mean_X) ** 2 for x in X) / M
empirical_var_Y = sum((y - empirical_mean_Y) ** 2 for y in Y) / M
empirical_cov_XY = sum((X[i] - empirical_mean_X) * (Y[i] - empirical_mean_Y) for i in range(M)) / M


plt.figure(figsize=(12, 6))
plt.hist(X, bins=50, density=True, alpha=0.6, label="X Density")
plt.hist(Y, bins=50, density=True, alpha=0.6, label="Y Density")
plt.title("Densities of Simulated X and Y")
plt.legend()
plt.show()

# Theoretical values
theoretical_mean_X = mu_X
theoretical_mean_Y = mu_Y
theoretical_var_X = var_X
theoretical_var_Y = var_Y
theoretical_cov_XY = cov_XY

results_with_theoretical = {
    "Empirical Mean X": empirical_mean_X,
    "Theoretical Mean X": theoretical_mean_X,
    "Empirical Mean Y": empirical_mean_Y,
    "Theoretical Mean Y": theoretical_mean_Y,
    "Empirical Variance X": empirical_var_X,
    "Theoretical Variance X": theoretical_var_X,
    "Empirical Variance Y": empirical_var_Y,
    "Theoretical Variance Y": theoretical_var_Y,
    "Empirical Covariance XY": empirical_cov_XY,
    "Theoretical Covariance XY": theoretical_cov_XY,
}

results_with_theoretical
```

## Steps:
1. Find the maximum value of \(f(x)\), either analytically or numerically.
2. Use the rejection method:
   - Generate \(Y ~ Uniform(0, 1)).
   - Generate \(U ~ Uniform(0, 1)).
   - Accept \(Y\) as \(X\) if:
     \[
     U <= f(Y)/c
     \]
3. Simulate \(M = 10,000\) samples.
4. Compute the empirical mean of accepted samples and compare it with the theoretical mean.

```python
import numpy as np

def f2_density(x):
    return 60 * (x**3) * ((1 - x)**2) if 0 < x < 1 else 0

def rejection_sampling(f, c, M):
    samples = []
    attempts = 0
    while len(samples) < M:
        Y = np.random.uniform(0, 1)
        U = np.random.uniform(0, 1)
        attempts += 1
        if U <= f(Y) / c:
            samples.append(Y)
    acceptance_rate = len(samples) / attempts
    return np.array(samples), acceptance_rate


def theoretical_mean():
    x_vals = np.linspace(0, 1, 1000)
    dx = x_vals[1] - x_vals[0]
    integral = sum(x * f2_density(x) * dx for x in x_vals)
    return integral


x_vals = np.linspace(0, 1, 1000)
c2 = max(f2_density(x) for x in x_vals)


M = 10000
samples2, acceptance_rate2 = rejection_sampling(f2_density, c2, M)


empirical_mean2 = np.mean(samples2)


theoretical_mean2 = theoretical_mean()

# Print results
print(f"Part 2: Acceptance Rate = {acceptance_rate2:.4f}")
print(f"Part 2: Empirical Mean = {empirical_mean2:.4f}")
print(f"Part 2: Theoretical Mean (same as empirical) = {theoretical_mean2:.4f}")
```

The max value of the function occurs at 1/4 with f(1/4) = 2.81 from the provided formulas

```python
def gamma(n):
    if n == 1:
        return 1
    elif n == 0.5:
        return 3.141592653589793**0.5
    else:
        return (n - 1) * gamma(n - 1)

def beta_density(x, alpha, beta):
    if 0 < x < 1:
        return (gamma(alpha + beta) / (gamma(alpha) * gamma(beta))) * (x**(alpha - 1)) * ((1 - x)**(beta - 1))
    else:
        return 0

def rejection_sampling_beta(alpha, beta, M, c):
    samples = []
    while len(samples) < M:
        Y = np.random.uniform(0, 1)
        U = np.random.uniform(0, 1)
        if U <= beta_density(Y, alpha, beta) / c:
            samples.append(Y)
    return samples

alpha, beta = 2, 4
M = 10000
x_vals = [i / 1000 for i in range(1, 1000)]
f_vals = [beta_density(x, alpha, beta) for x in x_vals]
c = max(f_vals)
samples = rejection_sampling_beta(alpha, beta, M, c)
empirical_mean = sum(samples) / len(samples)
empirical_variance = sum((x - empirical_mean)**2 for x in samples) / len(samples)
print(f"Part 1 (alpha=2, beta=4): Empirical Mean = {empirical_mean:.4f}, Empirical Variance = {empirical_variance:.4f}")

theoretical_mean = alpha / (alpha + beta)
theoretical_variance = (alpha * beta) / ((alpha + beta) ** 2 * (alpha + beta + 1))


print(f"Part 1 (alpha=2, beta=4): Theoretical Mean = {theoretical_mean:.4f}, Theoretical Variance = {theoretical_variance:.4f}")
```
